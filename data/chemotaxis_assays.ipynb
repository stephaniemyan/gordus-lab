```python
# colors = ["gold", "lightskyblue", "darkgoldenrod", "dodgerblue", "saddlebrown", "navy"]
colors = ["gold", "dodgerblue", "gold", "dodgerblue", "gold", "dodgerblue"]
sns.set_palette(colors)

labels = ["Parental (naive)", "Parental (trained)", "F1 (naive)", "F1 (trained)", "F2 (naive)", "F2 (trained)"]

def plot_boxplots(dataframe, time):
    
    df_plot = pd.DataFrame({"generation":dataframe.index.tolist()})
    
    df_plot["p0_o_CIs"] = dataframe["naive_choice_index"].tolist()
    df_plot["p0_p_CIs"] = dataframe["trained_choice_index"].tolist()
    df_plot["f1_o_CIs"] = dataframe["F1_naive_choice_index"].tolist()
    df_plot["f1_p_CIs"] = dataframe["F1_trained_choice_index"].tolist()
    df_plot["f2_o_CIs"] = dataframe["F2_naive_choice_index"].tolist()
    df_plot["f2_p_CIs"] = dataframe["F2_trained_choice_index"].tolist()
   
    df_plot.set_index("generation", inplace=True)
    
    # I checked, they're all normally distributed
#     print(scipy.stats.shapiro(df_plot['p0_o_CIs'].dropna()))
#     print(scipy.stats.shapiro(df_plot['p0_p_CIs'].dropna()))
#     print(scipy.stats.shapiro(df_plot['f1_o_CIs'].dropna()))
#     print(scipy.stats.shapiro(df_plot['f1_p_CIs'].dropna()))
#     print(scipy.stats.shapiro(df_plot['f2_o_CIs'].dropna()))
#     print(scipy.stats.shapiro(df_plot['f2_p_CIs'].dropna()))

    print(scipy.stats.ttest_ind(df_plot['p0_o_CIs'].dropna(),df_plot['p0_p_CIs'].dropna()))
    print(scipy.stats.ttest_ind(df_plot['f1_o_CIs'].dropna(),df_plot['f1_p_CIs'].dropna()))
    print(scipy.stats.ttest_ind(df_plot['f2_o_CIs'].dropna(),df_plot['f2_p_CIs'].dropna()))
    
    x1, x2 = 0, 1
    x3, x4 = 2, 3
    x5, x6 = 4, 5 # columns 'Sat' and 'Sun' (first column: 0, see plt.xticks())
    # y, h, col = df_plot["p0_o_CIs"].max() + 0.2, 0.2, 'k'
    
    fig, ax = plt.subplots(figsize=(15,7.5))
    sns.boxplot(data = df_plot)
    sns.swarmplot(data = df_plot, color = ".2")
    ax.set_title("Food choice after " + time + " training", fontsize=20)
    ax.set_ylabel("Food choice", fontsize=20)
    ax.set_xticklabels(labels, fontsize=15)
    ax.set_ylim([-1,1.25])
    plt.plot([x1, x1, x2, x2], [1.01, 1.1, 1.1, 1.01], lw=1, c='k')
    plt.plot([x3, x3, x4, x4], [1.01, 1.1, 1.1, 1.01], lw=1, c='k')
    plt.plot([x5, x5, x6, x6], [1.01, 1.1, 1.1, 1.01], lw=1, c='k')

    plt.text((x1+x2)*.5, 1.13, "***", ha='center', va='bottom', color='k')
    plt.text((x3+x4)*.5, 1.13, "n.s.", ha='center', va='bottom', color='k')
    plt.text((x5+x6)*.5, 1.13, "n.s.", ha='center', va='bottom', color='k')
    fig.savefig("/Users/cmdb/Documents/gordus-lab/data/" + time + ".png", transparent=True)
```

```python
plot_boxplots(df8, "8hr")
```

```python
plot_boxplots(df24, "24hr")
```
